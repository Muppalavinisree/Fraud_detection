Outlier detection and handling

```python
import pandas as pd
def detect_outliers_iqr(df):
    outlier_counts = {}
    for col in df.select_dtypes(include=['int64', 'float64']).columns:
        if df[col].nunique() <= 2:  
            continue  

        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower = Q1 - 1.5 * IQR
        upper = Q3 + 1.5 * IQR

        outliers = df[(df[col] < lower) | (df[col] > upper)]
        outlier_counts[col] = len(outliers)
    return outlier_counts
outlier_summary = detect_outliers_iqr(df)
print("Number of outliers per column:")
print(outlier_summary)
```

 Feature engineering

```python
import numpy as np
df["day"]  = (df["step"] // 24).astype(int)
df["hour"] = (df["step"] % 24).astype(int)
df["dest_is_merchant"] = df["nameDest"].str.startswith("M").astype(int)
for c in ["oldbalanceOrg","newbalanceOrig","oldbalanceDest","newbalanceDest"]:
    df[f"{c}_is_zero"] = (df[c] == 0).astype(int)
df["delta_org"]  = df["oldbalanceOrg"]  - df["newbalanceOrig"]
df["delta_dest"] = df["newbalanceDest"] - df["oldbalanceDest"]
df["err_org"]  = df["newbalanceOrig"]  + df["amount"] - df["oldbalanceOrg"]
df["err_dest"] = df["oldbalanceDest"] + df["amount"] - df["newbalanceDest"]
df["amt_to_oldbal"]   = df["amount"] / (df["oldbalanceOrg"]  + 1)
df["amt_to_destold"]  = df["amount"] / (df["oldbalanceDest"] + 1)
```

Multi-collinearity check and dropping highly collinear features

```python
import numpy as np
from sklearn.linear_model import Lasso
X = df.drop(['isFraud', 'isFlaggedFraud'], axis=1, errors='ignore').select_dtypes(include=[np.number]).fillna(0)
y = df['isFraud']
nunique = X.nunique()
constant_cols = nunique[nunique <= 1].index.tolist()
X = X.drop(columns=constant_cols)
print("Dropped constant columns:", constant_cols)
corr_matrix = X.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop_corr = [col for col in upper.columns if any(upper[col] > 0.95)]
X = X.drop(columns=to_drop_corr)
print("Dropped highly correlated features:", to_drop_corr)

corr_with_y = X.apply(lambda x: x.corr(y))
low_corr_cols = corr_with_y[abs(corr_with_y) < 0.1].index.tolist()
X = X.drop(columns=low_corr_cols)
print("Dropped low-correlation with target:", low_corr_cols)
lasso = Lasso(alpha=0.001, max_iter=5000)  
lasso.fit(X, y)

selected_features = X.columns[lasso.coef_ != 0].tolist()
X_clean = X[selected_features]

print("\nRemaining features after Lasso selection:", selected_features)
```

```python
X_clean = df[['delta_org', 'amt_to_destold']]
y = df['isFraud']
```

Train-test split and scaling

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(
    X_clean, y, test_size=0.2, random_state=42, stratify=y
)

print("Training samples:", X_train.shape[0])
print("Testing samples:", X_test.shape[0])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled, columns=X_train.columns, index=X_train.index)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=X_test.columns, index=X_test.index)

print("Feature scaling complete.")
```

Model building and selection (baseline + ensemble) and Evaluation (important metrics for imbalanced fraud problem)


```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from sklearn.metrics import classification_report, confusion_matrix, f1_score, roc_auc_score, precision_recall_curve, auc,average_precision_score
models = {
    "Logistic Regression": LogisticRegression(
        max_iter=1000, class_weight='balanced', random_state=42),
    "Random Forest": RandomForestClassifier(
        n_estimators=200, class_weight='balanced', random_state=42, n_jobs=-1),
    "XGBoost": xgb.XGBClassifier(
        n_estimators=200,scale_pos_weight=(y_train==0).sum()/(y_train==1).sum(),use_label_encoder=False,eval_metric='logloss',random_state=42,n_jobs=-1 )
}
for name, model in models.items():
    if name == "Logistic Regression":
        X_tr, X_te = X_train_scaled, X_test_scaled
    else:
        X_tr, X_te = X_train, X_test

    model.fit(X_tr, y_train)
    
    y_proba = model.predict_proba(X_te)[:,1] 
  
    
    roc_auc = roc_auc_score(y_test, y_proba)
    precision, recall, thresholds= precision_recall_curve(y_test, y_proba)
    f1=2*(precision*recall)/(precision+recall+1e-8)
    best_idx=np.argmax(f1)
    best_threshold=thresholds[best_idx]
    pr_auc = average_precision_score(y_test,y_proba)
    threshold=best_threshold
    y_pred=(y_proba>=threshold).astype(int)

    print(f"\n{name} Evaluation:")
    print(classification_report(y_test, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("F1 Score:", f1_score(y_test, y_pred))
    print("ROC-AUC:", roc_auc)
    print("Precision-Recall AUC:", pr_auc)
```
